```python
 
vmin, vmax = 0, 1874

# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(14,6))

# remove the axis
ax.axis('off')

# add a title and annotation
ax.set_title('Number of Covid Deaths in Flordia', fontdict={'fontsize': '25', 'fontweight' : '3'})

ax.set_xlim([-88, -80])
ax.set_ylim([25, 31.5])


# Create colorbar legend
sm = plt.cm.ScalarMappable(cmap='GnBu', norm=plt.Normalize(vmin=vmin, vmax=vmax))

# empty array for the data range
sm.set_array([])

cbar = fig.colorbar(sm)
#cbar.set_ticks([.2,.3, .4, .5, .6])
#cbar.set_ticklabels(['20%', '30%', '40%', '50%','60%'])
# create map
merged2.plot(column='deaths', cmap='GnBu', linewidth=1, ax=ax, edgecolor='.9')



# Add Labels
merged2['coords'] = merged2['geometry'].apply(lambda x: x.representative_point().coords[:])
merged2['coords'] = [coords[0] for coords in merged2['coords']]

# Display names 
for idx, row in merged2.iterrows():
    plt.annotate(s=row['deaths'], xy=row['coords'],horizontalalignment='center',fontsize=9)
```
